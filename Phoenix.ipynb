```python
from phoenix.otel import register
from opentelemetry.trace import Status, StatusCode
```

```python
phoenix_project_name = "example-rag-pipeline"
    # 在 Phoenix UI 介面上的資料夾名稱, 無責建立, 有則使用該資料夾
endpoint="http://" # url 端口
tracer_provider_phoenix = register(project_name=phoenix_project_name, endpoint = endpoint)
    # register(資料夾名稱, 接收端口) 註冊並建立 Tracer Provider 作為建立在本地, 連接收集數據與主機 (Phoenix) 的數據收發中心
tracer = tracer_provider_phoenix.get_tracer(__name__) # 取得 tracer, 在 trace 才能呼叫來記錄數據
```

```python
# example for "@tracer.chain"
# 取代在 Trace.ipynb 的 code base 用法, 但需要更細緻還是要用 code base
@tracer.chain
def format_documents(retrieved_docs):
    t = ''
    for i, doc in enumerate(retrieved_docs):
        t += f'Retrieved doc: {doc}\n'
    return t
```
